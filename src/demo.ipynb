```python
from pathlib import Path
import pandas as pd
import json


output_path=Path.cwd().parents[1].joinpath('Output','Companies')


df_full = pd.DataFrame()

for company_dir in output_path.iterdir():
    if company_dir.is_dir():
        crawled_file = company_dir / "crawled.json"
        if crawled_file.is_file():
            try:
                df=pd.json_normalize(json.load(open(crawled_file)))
                df_full = pd.concat([df_full, df],axis=0, ignore_index=True)
            except Exception as e:
                print(f'Error {str(e)} loading {crawled_file}')
            

df_full['length_url'] = df_full['url'].apply(lambda x: len(x))
df_full['lenght_html_string'] = df_full['html_string'].apply(lambda x: len(x))
df_full['html_lang'] = df_full['html_lang'].apply(lambda x: len(x))

df_full
```
